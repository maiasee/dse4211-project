# Sensitivity Analysis
Grid search over `lambda` and `max_weight` for both baseline and regime MVO portfolios.

```python
import pandas as pd
import numpy as np

from data_processing import load_data
from backtest import run_backtest

mu_base, mu_regime, real_returns, test_dates = load_data("outputs/outputs_draft/all_forecasts.csv")
```

```python
def compute_metrics(returns):
    sharpe = (returns.mean() / returns.std()) * (52 ** 0.5)
    cumulative_return = (1 + returns).prod() - 1
    cumulative = (1 + returns).cumprod()
    drawdown = cumulative / cumulative.cummax() - 1
    max_drawdown = drawdown.min()
    return sharpe, cumulative_return, max_drawdown
```

```python
lambdas = [1, 2, 5, 10, 20]
max_weights = [0.2, 0.3, 0.5]

rows = []

for lamda in lambdas:
    for max_weight in max_weights:
        for model_name, mu in [("baseline", mu_base), ("regime", mu_regime)]:
            ret, _ = run_backtest(
                mu, real_returns,
                lamda=lamda,
                max_weight=max_weight,
                eval_dates=test_dates
            )
            sharpe, cum_ret, max_dd = compute_metrics(ret)
            rows.append({
                "lambda": lamda,
                "max_weight": max_weight,
                "model": model_name,
                "sharpe": sharpe,
                "cumulative_return": cum_ret,
                "max_drawdown": max_dd,
            })
            print(f"lambda={lamda}, max_weight={max_weight}, model={model_name} — done")
```

```python
results = pd.DataFrame(rows, columns=["lambda", "max_weight", "model", "sharpe", "cumulative_return", "max_drawdown"])

print("Sensitivity Analysis Summary:")
results
```
